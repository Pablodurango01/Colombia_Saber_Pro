```python
# 01 - exploración.ipynb

import pandas as pd

# Cargar el archivo
df = pd.read_csv('train.csv')

# Vista general del dataset
print("Primeras filas del archivo:")
display(df.head())

print("\nInformación general del dataset:")
df.info()

print("\nResumen estadístico de columnas numéricas:")
display(df.describe())

# Columnas categóricas comunes que podrías explorar
columnas_interes = ['clase', 'localidad', 'edad', 'genero', 'estrato', 'institucion']

for columna in columnas_interes:
    if columna in df.columns:
        print(f"\n--- Análisis de la columna: '{columna}' ---")
        print(f"Valores únicos: {df[columna].nunique()}")
        print("Conteo de cada categoría:")
        print(df[columna].value_counts())
        print("-" * 40)

# Verificar si hay datos faltantes
print("\nCantidad de datos faltantes por columna:")
print(df.isnull().sum())
```
